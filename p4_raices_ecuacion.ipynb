## Utilizando la libreria Spicy.optmice encontrar todas las raices de la ecuacion

$$f(x) = x^5 + 5.25x^4 + 4.125x^3 - 9.125x^2 - 14.625x - 5.625$$

**Derivada de f(x)**

$$f'(x) = 5x^4 + 21x^3 + 12.375^2 - 18.25x^2 - 14.625$$


```python
## Utilizando la libreria Spicy.optmice encontrar todas las raices de la ecuacion $$f(x) = x^5 + 5.25x^4 + 4.125x^3 - 9.125x^2 - 14.625x - 5.625$$ **Derivada de f(x)** # $$f'(x) = 5x^4 + 21x^3 + 12.375^2 - 18.25x^2 - 14.625$$

from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)

# Definir la funcion f(x) 

def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

# Definir su derivada f'(x)
def df(x):
    return 5*x**4 + 21*x**3 + 12.375*x**2 - 18.25*x - 14.625

# Encontrar las raices utilizando diferentes puntos iniciales
initial_guesses = [-6, -4, -2, 0, 2]
roots = set()
for guess in initial_guesses:
    root = optimize.root_scalar(f, x0=guess, fprime=df)
    if root.converged:
        roots.add(round(root.root, 5))
# Mostrar las raices encontradas
print("Raices encontradas:")
for r in sorted(roots):
    print(r)

     
```
